```python
import tensorflow as tf
import numpy as np
```

```python
import csv
```

```python
X_train = []
Y_train = []

with open("train.csv",newline="") as f :

    reader = csv.reader(f)
    next(reader)
    i = 0
    for row in reader:
        X_train.append(list(map(int,row[1:])))
        Y_train.append(int(row[0]))
        i+=1

X= np.array(X_train)
Y = np.array(Y_train)
```

```python
model.fit(
    X,Y,epochs = 500
)
```

```python
predictions = model.predict(X)
```

```python

def max_index(A):
    maxi = 0
    for i in range(len(A)):
        if A[i] > A[maxi]:
            maxi = i
    return maxi  

yhat = [max_index(row) for row in predictions]
```

```python
wrong_i = []
wrong_x = []
wrong_y = []

for i in range(len(yhat)) :

    if (yhat[i] != Y[i]) :
        wrong_i.append(i)
        wrong_x.append(X[i])
        wrong_y.append(list([Y[i],yhat[i]]))
```

```python
print([(wrong_i[i],wrong_y[i]) for i in range(len(wrong_i))])
print(f"Right now total {len(wrong_i)} wrong predictions out of 40000 after 100 Epochs")
```

### Computing Cost

```python
def compute_cost(predictions,Y) :
    cost = 0
    ez = tf.nn.softmax(predictions)
    for i in range(len(Y)):
        # print(predictions[i][Y[i]])
        cost -= np.log(ez[i][Y[i]])

    return cost
```

```python
print(compute_cost(predictions,Y))
```
